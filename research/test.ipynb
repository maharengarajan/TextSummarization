```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv
from langchain import PromptTemplate
from langchain.chains.summarize import load_summarize_chain
```

```python
loader = PyPDFLoader("RFP.pdf")
docs = loader.load_and_split()
final_documents=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200).split_documents(docs)
```

```python
load_dotenv()
GROQ_API_KEY=os.getenv("GROQ_API_KEY")
llm=ChatGroq(api_key=GROQ_API_KEY,model="Llama3-8b-8192")
```

```python
chunks_prompt="""
Please summarize the below documents:
documents:`{text}'
Summary:
"""
map_prompt_template=PromptTemplate(input_variables=['text'],template=chunks_prompt)
```

```python
final_prompt='''
Provide the final summary of the entire documents with these important points.
Add a Title,Start the precise summary with an introduction and provide the summary in number 
points for the documents.
documents:{text}

'''
final_prompt_template=PromptTemplate(input_variables=['text'],template=final_prompt)
final_prompt_template
```

```python
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True,
)

output=summary_chain.run(final_documents)
output
```
